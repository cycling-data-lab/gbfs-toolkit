# gbfs-toolkit quickstart

A fully offline, reproducible walkthrough on the bundled sample feed. It needs only the core install (`pip install gbfs-toolkit`) and no network access, so a reviewer can run every cell end to end.

The sample is a small, deterministic GBFS snapshot of central Paris, parsed through the same `to_canonical_*` path as production data.

```python
import gbfs_toolkit as gb

gb.show_versions()
```

## 1. Load the bundled sample

`load_example` returns the canonical `(station_info, station_status)` pair.

```python
info, status = gb.load_example()
info
```

## 2. Run the static semantic audit (A1 to A7)

One row per station, with a boolean per rule, a `flagged` column, and a human-readable `reason`.

```python
verdict = gb.audit_static(info)
verdict[["station_id", "flagged", "reason"]]
```

## 3. Keep the trustworthy subset and compute occupancy

```python
clean = info.gbfs.drop_flagged()
av = gb.join_availability(info, status)
av["occupancy"] = gb.occupancy(av)
av[["station_id", "num_bikes_available", "num_docks_available", "occupancy"]]
```

## 4. A descriptive snapshot profile

```python
gb.system_profile(av)
```
